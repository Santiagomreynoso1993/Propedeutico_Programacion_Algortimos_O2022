## Objetivo
El objetio de esta pr√°ctia consiste en implementar el algoritmo Otsu para segmentaci√≥n binaria de una im√°gen en escala de grises.

## Implementaci√≥n:
Primero, importa los m√≥dulos necesarios.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

Ahora, carga la imagen 'cuadrados.png', convi√©rtela a escala de grises y mu√©strala.

```python
Cuadrados=cv.imread('cuadrados.png')
Cuadrados_gray=cv.cvtColor(Cuadrados,cv.COLOR_BGR2GRAY)
cv2_imshow(Cuadrados_gray)
```

Calcula el histograma de intensidad de la imagen y mu√©stralo

```python
hist=cv.calcHist([Cuadrados_gray],[0],None,[256],[0,255])
plt.plot(hist)
plt.show()
```

Implementa el algoritmo de Otsu descrito a continuaci√≥n.


Para cada umbral k, desde el umbral k=0 hasta k=255


      1.   Calcular W_A(k)
      2.   Calcular W_B(k)
      3.   Calcular Œº_A(k)
      4.   Calcular Œº_B(k)
      5.   Calcular œÉ2_A(K)
      6.   Calcular œÉ2_B(K)  
      7.   Calcular œÉ2_D(K)


Algoritmo de Otsu.jpg

```python
hist = hist.ravel()

best_threshold = 0
min_within_class_variance = float('inf')
within_class_variance=[]

for k in range(256):
  w_ak = np.sum(hist[:k+1])
  w_bk = np.sum(hist[k+1:])

  if w_ak == 0 or w_bk == 0:
    within_class_variance.append(float('inf'))
    continue
  m_ak =np.sum(np.arange(0, k+1)*hist[:k+1])/w_ak
  m_bk =np.sum(np.arange(k+1,256)*hist[k+1:])/w_bk

  sigma_ak_cuad = np.sum(((np.arange(0, k+1)-m_ak)**2)*hist[:k+1])/w_ak
  sigma_bk_cuad = np.sum(((np.arange(k+1,256)-m_bk)**2)*hist[k+1:])/w_bk

  sigma_dk_cuad = w_ak*sigma_ak_cuad + w_bk*sigma_bk_cuad

  within_class_variance.append(sigma_dk_cuad)

  if sigma_dk_cuad < min_within_class_variance:
    min_within_class_variance = sigma_dk_cuad
    best_threshold = k
```

Encuentra el umbral √≥ptimo k* que minimiza la varianza entre las clases œÉ2_D(K).

Imprime el valor del umbral √≥ptimo k*.

```python
print('Best threshold:', best_threshold)
```

Grafica la variaza entre las clases œÉ2_D(K) para cada umbral k. Adem√°s, grafica el punto m√≠nimo encontrado, el cual corresponde al umbr√°l √≥ptimo k*.

```python
plt.plot(within_class_variance, label = 'Varianza dentro de clases œÉ2_D(K)')
plt.axvline(x=best_threshold, color='r', linestyle='--', label = f'Umbral √≥ptimo k*: {best_threshold}')
plt.legend()
plt.xlabel('Umbral')
plt.ylabel('Varianza dentro de clases œÉ2_D(K)')
plt.show()
```

Binarizar la imagen utilizando el umbral √≥ptimo k*. Muestra la imagen segmentada con el umbral √≥ptimo.

```python
cuadrados_binarizados= cv.threshold(Cuadrados_gray, best_threshold, 255, cv.THRESH_BINARY)
plt.imshow(cuadrados_binarizados[1], cmap='gray')
plt.show()
```

Muestra el histograma de intensidad de la imagen en escala de grises y dibuja la l√≠nea que indica el umbral √≥ptimo k* encontrado.

```python
plt.figure(figsize=(20, 12))
plt.hist(Cuadrados_gray.ravel(), bins=256, range=(1, 255), color='gray', alpha=0.7, label='Histrograma de Intensidades')
plt.axvline(x=best_threshold, color='r', linestyle='--', linewidth= 3, label = f'Umbral √≥ptimo k*: {best_threshold}')
plt.legend()
plt.xlabel('Nivel de Intensidad')
plt.ylabel('Frecuencia')
plt.title('Histograma de intensidad de la imagen en escala de grises')
```
